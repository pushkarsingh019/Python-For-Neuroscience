```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Here we are plotting two lines.
# eqn for line 1: y1 = 0.3x + 30. 
# eqn for line 2: y2 = 0.5x + 10.

m1,c1 = [0.3, 30]
m2,c2 = [0.5,10]

x = np.arange(20,70,1)

y1 = m1*x + c1
y2 = m2*x + c2

plt.title("Plotting Two Lines")
plt.plot(x,y1,color='purple')
plt.plot(x,y2,color='hotpink')
plt.show()
```

```python
# Problem 1: Redoing the plot with the new slopes and intercepts.
m1,c1,m2,c2 = [0.05, 50, 1, 5]

y1= m1*x+c1
y2= m2*x+c2

# fig = plt.figure(figsize=(6,6))
plt.title(f"y = {m1}x + {c1}, y = {m2}x + {c2}")
plt.plot(x,y1,'-',color='red')
plt.plot(x,y2,'-',color='blue')
plt.show()
```

```python
# Problem 2: adding labels to the above plot.
fig = plt.figure(figsize=(6,6))
plt.plot(x,y1,'-',color='red', label="Female")
plt.plot(x,y2,'-',color='blue', label="Male")
plt.legend()
plt.xlabel('Age')
plt.ylabel('TDT (ms)')

plt.plot()
```
