```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting the following line: y = 0.3x + 10. from 20 to 70.
# slope = 0.3, y-intercept = 10.

m = 0.3
c = 10
x = np.arange(20,70,1)

y = m*x + c
plt.plot(x,y, '-', color='b')
plt.xlabel('Age')
plt.ylabel('Temporal Discrimination (ms)')
plt.show()
```

```python
# solution for problem 1
# Here we have to plot the same line but with slope = 1, intercept = -5

m = 1
c = -5
np.arange(20,70,1)

y = m*x+c

plt.title("Problem 1")
plt.plot(x,y,color='r')
plt.show()
```

```python
# solution for problem 2
# Here we have to redo the above plot with different color

m = 1
c = -5
np.arange(20,70,1)

y = m*x+c

plt.title("Problem 2")
plt.plot(x,y,color='purple')
plt.show()
```
