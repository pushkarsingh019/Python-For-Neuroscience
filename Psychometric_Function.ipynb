```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
stimulus_interval = np.arange(0,101,1)

pse = 50
jnd = 10
tdt_function = norm.cdf(stimulus_interval, pse, jnd)
```

```python
plt.plot(stimulus_interval, tdt_function, color="black")
plt.show()
```

```python
gaussian = norm.pdf(stimulus_interval, pse, jnd)
plt.plot(stimulus_interval, gaussian)
plt.show()
```

```python
# Problem 1: redo the plot with pse of 40 and jnd of 30.
PSE = 40
JND = 30
ISI = np.arange(0,101, 1)

psychometric = norm.cdf(ISI, PSE, JND)
plt.plot(psychometric)
plt.show()
```

```python
# Problem 2: different coloured curve
plt.plot(psychometric, color='magenta')
plt.show()
```

```python
# Problem 3: What is the probability that someone will see the two flashing lights if the interstimulus interval is 60ms apart given a PSE of 45 and JND of 15.
x_value = 60
PSE=45
JND=15
psychometric= norm.cdf(ISI,PSE,JND)
probability = norm.cdf(60, PSE, JND)
fig,ax = plt.figure(figsize=(4,4)),plt.gca()
plt.plot(ISI,psychometric,'-',color='blue')
plt.xlabel('Interstimulus Interval (ms)')
plt.ylabel('Proportion of Different Responses')
plt.axvline(x = 60,color='red', linestyle = '--', ymin=0, ymax = probability / psychometric[-1])
plt.axhline(probability, color='red', linestyle="--", xmin=0, xmax=60 / ISI[-1])
plt.plot(60,probability, 'ro')
plt.annotate(f'({x_value}, {probability:.2f})', 
             xy=(x_value, probability), 
             xytext=(x_value + 5, probability - 0.1))
plt.show()
```
